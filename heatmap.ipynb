# Running Heatmap Debug Notebook

This notebook keeps the VS Code workflow while using the same `heatmap_core.py` code as the local app. Edit the config cell, then run the generation cell.

```python
from pathlib import Path

from heatmap_core import HeatmapConfig

# Existing unzipped Strava export folder for VS Code debugging.
EXPORT_DIR = Path("strava_export")
OUTPUT_HTML = Path("outputs/heatmap.html")
CACHE_DIR = Path("cache")

CONFIG = HeatmapConfig(
    activity_types=["Run"],
    date_from=None,
    date_to=None,
    home_lat=None,
    home_lon=None,
    radius_km=None,
    track_clip_radius_km=None,
    meters_per_pixel=3,
    max_grid_pixels=250_000_000,
    blur_sigma_m=25,
    default_visible_layer="Raw GPS tracks",
)
```

```python
from IPython.display import IFrame, display

from heatmap_core import generate_heatmap

result = generate_heatmap(
    export_dir=EXPORT_DIR,
    output_html=OUTPUT_HTML,
    config=CONFIG,
    cache_dir=CACHE_DIR,
)

print(result)
display(IFrame(str(result.output_html), width="100%", height=760))
```
